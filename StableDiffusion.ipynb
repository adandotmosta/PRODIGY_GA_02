```python
import keras_cv
import keras
import matplotlib.pyplot as plt
```

```python
keras.mixed_precision.set_global_policy("mixed_float16")
model = keras_cv.models.StableDiffusion(jit_compile=True, img_height = 256, img_width = 256)
```

```python
images = model.text_to_image(" a man dancing with an umbrella under the sea "    "high quality, highly detailed, elegant, sharp focus, "
    "concept art, character concepts, digital painting", batch_size = 3)
```

```python
def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")


plot_images(images)
```
